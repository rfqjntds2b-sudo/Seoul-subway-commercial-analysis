```python
import pandas as pd
import numpy as np
import os
import re
from math import radians, cos, sin, asin, sqrt

# 1. 경로 설정
BASE_PATH = '/content/drive/MyDrive/Colab Notebooks/2026/이어드림스쿨6기/subway_project/'

print("--- 1. 데이터 로드 시작 ---")

# (1) 지하철 데이터 (UTF-8, 탭 분리) - UnicodeDecodeError 발생으로 인해 'euc-kr' 인코딩으로 변경
subway_raw = pd.read_csv(os.path.join(BASE_PATH, 'Seoul_subway_data_2023_2026.csv'), sep='\t', encoding='euc-kr', header=None)

# (2) 지하철 위치 데이터
subway_geo = pd.read_csv(os.path.join(BASE_PATH, 'subway_location_data_2023_2026.csv'), encoding='utf-8')

# (3) [업데이트] 상가 정보 데이터 로드 (DtypeWarning 방지를 위해 low_memory=False 추가)
shop_raw = pd.read_csv(os.path.join(BASE_PATH, 'shop.csv'), encoding='utf-8', low_memory=False)

# (4) 버스 위치 데이터 - UnicodeDecodeError 발생으로 인해 'euc-kr' 인코딩으로 변경
bus_geo = pd.read_csv(os.path.join(BASE_PATH, 'bus_location_data.csv'), encoding='euc-kr')

# (5) 연도별 버스 승하차 데이터 병합
bus_files = ['filtered_bus_data_2023.csv', 'filtered_bus_data_2024.csv', 'filtered_bus_data_2025.csv', 'filtered_bus_data_2026.csv']
bus_list = []
for file in bus_files:
    file_path = os.path.join(BASE_PATH, file)
    if os.path.exists(file_path):
        bus_list.append(pd.read_csv(file_path, encoding='utf-8'))
bus_raw = pd.concat(bus_list, ignore_index=True)

print("모든 원천 데이터가 성공적으로 로드되었습니다!")

print("\n--- 2. 데이터 전처리 및 타겟팅 ---")

# [지하철 전처리] 컬럼 생성 및 하차 인원 계산
col_names = ['사용월', '호선', '역명']
for hour in range(4, 28):
    h = hour if hour < 24 else hour - 24
    col_names.extend([f'{h:02d}시승차', f'{h:02d}시하차'])
subway_raw.columns = col_names[:len(subway_raw.columns)]

subway_alight_cols = [col for col in subway_raw.columns if '하차' in col]
subway_raw['지하철_총하차'] = subway_raw[subway_alight_cols].sum(axis=1)
subway_raw['정제역명'] = subway_raw['역명'].str.split('(').str[0].str.rstrip('역')

# 최신 2026년 5월 기준 데이터 취합
subway_202605 = subway_raw[subway_raw['사용월'] == 202605]
subway_summary = subway_202605.groupby('정제역명')['지하철_총하차'].sum().reset_index()

# 지하철 좌표 매칭 및 정제: '지하철역', 'y좌표', 'x좌표' 사용
subway_geo['정제역명'] = subway_geo['지하철역'].str.split('(').str[0].str.rstrip('역')
subway_geo_sub = subway_geo.groupby('정제역명')[['y좌표', 'x좌표']].mean().reset_index() # 중복 역 좌표 평균화
subway_master = pd.merge(subway_summary, subway_geo_sub, on='정제역명', how='inner')
subway_master = subway_master.rename(columns={'y좌표': '위도', 'x좌표': '경도'}) # 컬럼명 통일


# [버스 전처리] 버스 총 하차량 구하기
bus_alight_cols = [col for col in bus_raw.columns if '하차' in col or '승객' in col] # 데이터 컬럼명에 맞게 자동 선택
bus_raw['버스_총하차'] = bus_raw[bus_alight_cols].sum(axis=1)

# 버스 승하차 데이터의 '역명'에서 괄호 안의 숫자 제거하여 정류소명 정제
bus_raw['정류소명_정제'] = bus_raw['역명'].str.split('(').str[0].str.strip()

# 버스 위치 정보 데이터의 컬럼명 통일 (kernel state의 bus_geo에 맞춤)
# '노드 ID' -> '정류소ID', 'Y좌표' -> '위도', 'X좌표' -> '경도'
bus_geo_processed = bus_geo.rename(columns={'노드 ID': '정류소ID', 'Y좌표': '위도', 'X좌표': '경도'})

# 버스 승하차 데이터와 버스 위치 정보를 '정류소명'을 기준으로 결합
# 이 결합을 통해 bus_raw 데이터에 '정류소ID', '위도', '경도' 컬럼이 추가됩니다.
bus_raw_with_geo = pd.merge(bus_raw,
                            bus_geo_processed[['정류소ID', '정류소명', '위도', '경도']],
                            left_on='정류소명_정제',
                            right_on='정류소명',
                            how='left')

# 병합 후 '정류소명' 관련 중복 컬럼 제거 및 필요한 컬럼만 선택
bus_raw_with_geo.drop(columns=['정류소명_정제', '정류소명'], inplace=True)

# 정류소ID가 없는 (매칭 실패한) 데이터는 분석에서 제외
bus_raw_with_geo.dropna(subset=['정류소ID'], inplace=True)
# 정류소ID를 정수형으로 변환 (dropna 후 진행)
bus_raw_with_geo['정류소ID'] = bus_raw_with_geo['정류소ID'].astype(int)

# 버스 최신 데이터 정류소별 합산 후 위치(위경도) 데이터와 결합
# 이제 bus_raw_with_geo에 '정류소ID', '위도', '경도'가 있으므로 이를 기준으로 그룹바이
bus_summary = bus_raw_with_geo.groupby('정류소ID').agg({
    '버스_총하차': 'mean',
    '위도': 'mean',
    '경도': 'mean'
}).reset_index()

# 최종 버스 마스터 데이터는 bus_summary 자체가 됩니다.
bus_master = bus_summary.copy()


# ➒ [핵심 업데이트] 상권 데이터 매칭 필터링
# 대분류명이 '음식'인 데이터(주점, 식당 포함)만 정밀 추출합니다.
shop_food = shop_raw[shop_raw['상권업종대분류명'] == '음식'].copy()
# 결측치 제거 (좌표가 없는 행 방지)
shop_food = shop_food.dropna(subset=['위도', '경도'])


print("\n--- 3. 공간 결합 알고리즘 가동 (반경 500m 이내 스캔) ---")

# 하버사인 거리 계산 함수
def haversine_distance(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat, dlon = lat2 - lat1, lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    return 2 * asin(sqrt(a)) * 6371 * 1000 # 미터 단위 반환

subway_master['주변_버스_하차'] = 0
subway_master['주변_상가_수'] = 0

# 지하철역 좌표 기준으로 루프 연산
for s_idx, s_row in subway_master.iterrows():
    s_lat, s_lon = s_row['위도'], s_row['경도']

    # 1) 반경 500m 이내 버스 정류장 하차 인구 합산
    b_distances = bus_master.apply(lambda b: haversine_distance(s_lat, s_lon, b['위도'], b['경도']), axis=1)
    subway_master.at[s_idx, '주변_버스_하차'] = bus_master[b_distances <= 500]['버스_총하차'].sum()

    # 2) 반경 500m 이내 상가(음식) 개수 카운트
    # 원본 파일 컬럼명인 '위도', '경도'를 정확히 참조합니다.
    s_distances = shop_food.apply(lambda sh: haversine_distance(s_lat, s_lon, sh['위도'], sh['경도']), axis=1)
    subway_master.at[s_idx, '주변_상가_수'] = (s_distances <= 500).sum()

print("지하철역 주변 대중교통 인구 및 상가 인프라 공간 매칭 완료!")

print("\n--- 4. 플랫폼용 '혼잡 가성비(C/P) 지수' 연산 ---")

# 대중교통 통합 유입량
subway_master['총_유입_인구'] = subway_master['지하철_총하차'] + subway_master['주변_버스_하차']

# 가성비 지수 수식 (상가 수 / 총 유입 인구 * 10,000)
subway_master['혼잡_가성비_지수'] = (subway_master['주변_상가_수'] / (subway_master['총_유입_인구'] + 1)) * 10000

# 지수 기준 내림차순 정렬 (가성비 좋은 '숨은 꿀 상권'이 최상단에 올라옴)
final_report_df = subway_master.sort_values(by='혼잡_가성비_지수', ascending=False)

# 최종 마스터 CSV 파일 내보내기 (한글 깨짐 방지 utf-8-sig)
output_file = os.path.join(BASE_PATH, 'subway_bus_shop_merged_result.csv')
final_report_df.to_csv(output_file, index=False, encoding='utf-8-sig')

print(f"🎉 네이버/카카오 지도 플랫폼 제안용 마스터 데이터셋 구축 완료!\n저장 경로: {output_file}")
final_report_df[['정제역명', '지하철_총하차', '주변_버스_하차', '주변_상가_수', '혼잡_가성비_지수']].head(10)
```
